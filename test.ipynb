```python
from pydub import AudioSegment

def pitch_shift(audio, semitones):
    """
    semitones: 올릴 음정 (양수면 올림, 음수면 내림)
    1 semitone = 반음 (C→C#)
    """
    # 음정 변경 비율 계산
    pitch_ratio = 2 ** (semitones / 12)
    
    # 프레임레이트 조정으로 음정 변경
    new_sample_rate = int(audio.frame_rate * pitch_ratio)
    
    # 새로운 오디오 생성 후 원래 레이트로 재설정
    shifted = audio._spawn(audio.raw_data, overrides={'frame_rate': new_sample_rate})
    return shifted.set_frame_rate(audio.frame_rate)

# 사용 예제
sound1 = AudioSegment.from_file("C:/Users/user/Downloads/화개장터_아이유+4.wav")
sound2 = AudioSegment.from_file("C:/Users/user/Downloads/화개장터_mr.mp3")

# sound2를 5 반음 올리기 (2음 올림)
sound2_shifted = pitch_shift(sound2, 4)

# 겹치기
combined = sound1.overlay(sound2_shifted)

# 저장
combined.export("C:/Users/user/Downloads/화개장터_아이유+4_.wav", format='wav')
```

```python
from spleeter.separator import Separator
from spleeter.audio.adapter import AudioAdapter
import os

input_audio = "C:/Users/user/Documents/카카오톡 받은 파일/gichan_voice/combined/"  # mp3, wav, m4a 등 지원
base_name = os.path.splitext(input_audio)[0]

output_dir = "C:/Users/user/Downloads"
os.makedirs(output_dir, exist_ok=True)

separator = Separator('spleeter:2stems')
audio_loader = AudioAdapter.default()

# sample_rate 인자를 빼면 파일에 맞는 rate로 자동 로드됨
audio, sample_rate = audio_loader.load(input_audio)

prediction = separator.separate(audio)

vocal_path = os.path.join(output_dir, f"{base_name}_mr제거.wav")
mr_path = os.path.join(output_dir, f"{base_name}_mr.wav")

audio_loader.save(vocal_path, prediction['vocals'], sample_rate)
audio_loader.save(mr_path, prediction['accompaniment'], sample_rate)

print(f"보컬 파일: {vocal_path}")
print(f"반주 파일: {mr_path}")
```

```python
import os
from spleeter.separator import Separator
from spleeter.audio.adapter import AudioAdapter
from pydub import AudioSegment

folder = "C:/Users/user/Downloads/test"
audio_extensions = {'.mp3', '.wav', '.m4a', '.flac', '.ogg', '.aac'}

separator = Separator('spleeter:2stems')
audio_loader = AudioAdapter.default()

vocal_segments = []

for file in os.listdir(folder):
    base, ext = os.path.splitext(file)
    if ext.lower() in audio_extensions:
        print(f"처리 중: {file}")

        audio, sample_rate = audio_loader.load(os.path.join(folder, file))
        prediction = separator.separate(audio)

        # 임시로 보컬만 저장 (필요 시)
        vocal_path = os.path.join(folder, f"{base}_vocal.wav")
        audio_loader.save(vocal_path, prediction['vocals'], sample_rate)

        # 보컬 오디오 로드 (PyDub)
        vocal_audio = AudioSegment.from_file(vocal_path)
        vocal_segments.append(vocal_audio)

        # 임시 파일 삭제 원하면 아래 주석 해제
        # os.remove(vocal_path)

# 여러 보컬 파일을 차례대로 이어붙이기
combined_vocals = sum(vocal_segments)

# 합친 보컬 wav 저장
output_path = os.path.join(folder, "combined_vocals.wav")
combined_vocals.export(output_path, format='wav')

print(f"모든 보컬 파일 합쳐서 저장 완료: {output_path}")
```
